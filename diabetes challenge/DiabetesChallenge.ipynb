```python
X_train = dfx.values
```

```python
Y_train = dfy.values
```

```python
test_df = pd.read_csv("C:/diabetes-classification-challenge/Dataset/TestData/Diabetes_Xtest.csv")
```

```python
X_test = test_df.values
```

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
nearestpoints = 17
def knn(X,Y,test,k=nearestpoints):
    vals = []
    m = X.shape[0]
    
    for i in range(m):
        d = dist(test,X[i])
        vals.append((d,Y[i]))
        
    vals = sorted(vals)
    vals = vals[:k]
    vals = np.array(vals)
    
    new_vals = np.unique(vals[:,1],return_counts=True)
    index = new_vals[1].argmax()
    pred = new_vals[0][index]
    return(pred)
```

```python
for i in range(X_test.shape[0]):
    pred = knn(X_train,Y_train,X_test[i])
    outcome.append(int(pred))
```

```python
outcome = np.asarray(outcome)
df = pd.DataFrame(outcome)
```

```python
df.to_csv("C:/diabetes-classification-challenge/submission17"+str(nearestpoints)+".csv",index=False,header=0)
```
